    Checking Length
    Delete First
    Delete Last
    Insert First
    Insert Last
    List Size
    Traversal
    

```python
class SLList:
    def __init__(self): 
        self.head = None
    
    def insertFirst(self, new_data): 
  
        new_node = Node(new_data) 
  
        new_node.next = self.head 
  
        self.head = new_node 

    def insertLast(self, new_data): 

        new_node = Node(new_data) 

        if self.head == None:
            self.head = new_node 
            return

        last = self.head 
        while (last.next): 
            last = last.next
        last.next =  new_node 
    
    def insertAfter(self, target_node_data, new_data): 
  
        if not self.listSize():
            print("List is Empty")
            return

        currentNode = self.head
        new_node = Node(new_data)
        while(currentNode != None):
            if currentNode.element == target_node_data:
                new_node.next = currentNode.next
                currentNode.next = new_node
                return True
            else:
                currentNode = currentNode.next
            

        else:
            print("Search Element NOT in the List")

        
    def listSize(self): 
        temp = self.head # Initialise temp 
        count = 0 # Initialise count 
  
        # Loop while end of linked list is not reached 
        while (temp): 
            count += 1
            temp = temp.next
        return count 
        
#@end-editable@
    
    def printList(self):
        if (self.isEmpty()):
            print("List is Empty")
        else:
            currentNode = self.head
            while(currentNode != None):
                print(currentNode.element, end =" ")
                currentNode = currentNode.next
            print("")
    
    def isEmpty(self):
        if self.listSize():
            return False
        else:
            return True

    def deleteFirst(self):
        temp = self.head
        self.head = temp.next
        temp = None
        return 

    def deleteLast(self):
        for i in range(self.listSize() -1 ): 
            temp = temp.next
            if temp is None: 
                break
  
#@start-editable@




class Node:
    def __init__(self, element=None):
        self.element= element
        self.next = None

```

```python
def testSLL():
    sll = SLList()
    inputs=int(input())
    while inputs>0:
        command=input()
        operation=command.split()
        if(operation[0]=="S"):
        	print(sll.listSize())
        elif(operation[0]=="I"):
        	print(sll.isEmpty())
        elif(operation[0]=="IF"):
            sll.insertFirst(int(operation[1]))
            sll.printList()
        elif(operation[0]=="IL"):
            sll.insertLast(int(operation[1]))
            sll.printList()
        elif(operation[0]=="DF"):
            sll.deleteFirst()
            sll.printList()
        elif(operation[0]=="DL"):
            sll.deleteLast()
            sll.printList()
        elif(operation[0]=="IA"):
            returnVal = sll.insertAfter(int(operation[1]), int(operation[2]))
            if(returnVal == True):
                sll.printList()
        inputs-=1

testSLL()

```
